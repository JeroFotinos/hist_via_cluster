- DESCR, a string with a description of the dataset, just leave it as an empty string for now.
- diet, a one-dimensional numpy array that contains integers corresponding to the mapping: {0 : 'control', 1 : 'omega3', 2 : 'omega6'}
- diet_names, a list of strings with the diet names for each number, i.e., ['control', 'omega3', 'omega6']
- mouse, a one-dimensional numpy array that contains integers corresponding to the raton number
- take, a one-dimensional numpy array that contains integers corresponding to the toma number
- images, a four-dimensional numpy array that contains the 3D images for each diet, mouse, and take. E.g., images[n] contains the three-dimensional np.array that corresponds to diet[n], mouse[n], and take[n]. The three dimensional images should be the agregation of the 2D images in the order specified by element_order. That is, images[n][0] is the 2D image for the fluorescence of Ca, of mouse[n], take[n] with diet[n].
- element_order, a list of strings with the names of the element in the order used to build the 3D images for each mouse, take, and diet, i.e., ['Ca', 'Cu', 'Fe', 'K', 'Mn', 'P', 'S', 'Zn'].
